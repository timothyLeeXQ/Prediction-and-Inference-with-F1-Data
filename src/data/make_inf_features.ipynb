## Table generation

Filter to include only races from 1950 - 2017 (inclusive)

Needed references to cross reference other tables and build features:
* Race ID
* Driver ID
* Constructor ID

Outcomes
* Outcome 1 - First or Second
* Outcome 2 - positionOrder (remember to filter for 1 or 2 when running analysis)

Base Features
* Grid Position

Features to generate/join
* Driver skill - Years since debut
* Number of pit stops
* Constructor quality - whether the constructor has won a championship
* Circuit

Interactions
* Driver skill x grid position
* Driver x track
* Driver x constructor
* Pit stops x driver

```
df_refs_with_outcome = df_results.select('raceId', 'driverId', 'constructorId', 'grid', 'positionOrder')\
  .withColumn('1_or_2',
              F.when(F.col('positionOrder') > 2, 'lose')\
              .otherwise('win'))\
  .join(df_races.select('raceId', 'year'), on = ['raceId'], how = 'left')\
  .filter(df_refs_with_outcome['year'] >= 1950)\
  .filter(df_refs_with_outcome['year'] <= 2017)
              
display(df_refs_with_outcome)
```

```
df_refs_with_outcome.count()
```

```
#Use to check for missing values in positionOrder and grid
# display(df_refs_with_outcome.groupby('grid').count())
# display(df_refs_with_outcome.groupby('grid').count().agg(F.sum("count")))
# display(df_refs_with_outcome.groupby('positionOrder').count())
display(df_refs_with_outcome.groupby('positionOrder').count().agg(F.sum("count")))
```

There are missing values in grid position, labelled 0

These are probably DNQs or crashes during qualifying. 
* All but a handful have unrealistic finishing positions of high 20s or 30s,
suggesting that these positions were inferred.
* None of these position 0s had a podium finish, while only 11 finished in 
the top 15.

I think it's safe to remove these values from the analysis.

```
df_refs_with_outcome = df_refs_with_outcome.filter(df_refs_with_outcome['grid'] != 0)
display(df_refs_with_outcome)
```

```
df_refs_with_outcome.count()
#1566 missing values removed
```

## Driver Skill

```
driver_debut_year = df_refs_with_outcome\
  .groupby('driverId')\
  .agg(F.min(F.col('year')).alias('debut_year'))
display(driver_debut_year)
```

```
df_refs_outcome_driver_exp = df_refs_with_outcome.join(driver_debut_year, on = ['driverId'], how = 'left')\
  .withColumn("years_since_debut", F.col('year') - F.col("debut_year"))\
  .drop('debut_year', 'year')
display(df_refs_outcome_driver_exp)
```

```
df_refs_outcome_driver_exp.count()
```

## Number of Pit Stops

```
pit_strat = df_pitstops.groupby('raceId', 'driverId').agg(F.max(F.col('stop')).alias('num_pits'))\
  .withColumn('pit_strategy',
              F.when(F.col('num_pits') > 3, '>3-stops')\
              .otherwise(F.col('num_pits')))\
  .select('raceId', 'driverId', 'pit_strategy')
display(pit_strat)
```

```
df_refs_outcome_driver_exp_pit = df_refs_outcome_driver_exp.join(pit_strat, on= ['raceId', 'driverId'], how = 'left')
display(df_refs_outcome_driver_exp_pit)
```

### Missing Values

```
df_refs_outcome_driver_exp.select('raceId').distinct().count()
```

We have a lot of null values here. This seems to be because 
we only have pitstop data for a minority of races
(see code just above)

Removing these null values leads to questions as to 
whether the pattern of missingless in races with 
available pitstop data is a random sample. 

Spoiler, it is not, given that the raceIds of data with pitstops 
ranges from 841 to 1030 (get this by sorting df pitstops), omitting 
1-840 and 1031-1040. These IDs mostly correspond to races
in the 2010s.

Imputation is a bad strategy for an inferential model, since 
we should understand what's happening by inference first before
trying to impute strategies used. Also, as there is a missingness
pattern that isn't random, imputation might result in a bias in the
pit strategy data.

I think it is best to leave null values as their own nominal class.
This could correspond to 'old strategies' of races before the 2010s.

```
df_refs_outcome_driver_exp_pit = df_refs_outcome_driver_exp_pit.fillna({'pit_strategy':'missing'})
display(df_refs_outcome_driver_exp_pit)
```
